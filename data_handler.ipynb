```python
import constants
from helpful_scripts import read_from_text_to_dicts_list, get_today_file_path
import re
from datetime import date
```

```python
def check_price_range(item_list, below_market_price, above_market_price):

    price_range = []

    for i in item_list:
        price = i['Price']
        price_range.append(int(price))
    
    min = 0
    max = 0
    min_post = None

    for  index, i in enumerate(price_range):
        
        if min == 0:
            min = i
            min_post = item_list[index]
        elif i < min and i > below_market_price:
            min = i
            min_post = item_list[index]
       
        if i > max and i < above_market_price:
            max = i

    return (min, max, min_post)
```

```python
data_file = get_today_file_path(constants.SMARTPHONES[0]['file'])
apple_iphone_and_ipad = read_from_text_to_dicts_list(data_file)
```

```python
def check_iphone_series():
    iphone13 = []
    iphone12 = []
    iphone14 = []
    iphone11 = []
    iphone10 = []
    below_ten = []
    ipad = []
    for i in apple_iphone_and_ipad:
        
        try:
            str_iphone = i["Name"].replace(" ", "")
            # print(f"name: {str_iphone}")
            series = str_iphone.lower().split("phone")
            if series[1][0] != 'x':
                series_str = series[1][0] + series[1][1]
            else:
                series_str = 'x'
        except Exception:
            try:
                str_iphone = i["Description"].replace(" ", "")
                # print(f"description: {str_iphone}")
                series = str_iphone.lower().split("phone")
                series_str = series[1][0] + series[1][1]
            except Exception:
                series_str = i["Name"]

        if "14" in series_str:
            iphone14.append(i)
        elif "13" in series_str:
            iphone13.append(i)
        elif "12" in series_str:
            iphone12.append(i)
        elif "11" in series_str:
            iphone11.append(i)
        elif "10" in series_str or "x" in series_str[0]:
            iphone10.append(i)
        elif "ipad" in i["Name"].lower() or "i pad" in i["Description"].lower():
            ipad.append(i)
        else:
            below_ten.append(i)

    return below_ten, iphone10, iphone11, iphone12, iphone13, iphone14, ipad




below_ten, iphone10, iphone11, iphone12, iphone13, iphone14, ipad = check_iphone_series()
print(f"Total Apple product post in Oct: {len(apple_iphone_and_ipad)}")

print(f"iPhone 14: {len(iphone14)}")
(iphone14_min, iphone14_max, min_post_iphone14) = check_price_range(iphone14, 500, 2500)
print(f"iPhone 14 Price Range: {iphone14_min}$ - {iphone14_max}$\n")
print("Min post link:")
print(min_post_iphone14["Post Link"])
print()

print(f"iPhone 13: {len(iphone13)}")
iphone13_min, iphone13_max, min_post_iphone13 = check_price_range(iphone13, 400, 2000)
print(f"iPhone 13 Price Range: {iphone13_min}$ - {iphone13_max}$\n")
print("Min post link:")
print(min_post_iphone13["Post Link"])
print()

print(f"iPhone 12: {len(iphone12)}")
iphone12_min, iphone12_max, min_post_iphone12 = check_price_range(iphone12, 200, 2000)
print(f"iPhone 12 Price Range: {iphone12_min}$ - {iphone12_max}$\n")
print("Min post link:")
print(min_post_iphone12["Post Link"])
print()

print(f"iPhone 11: {len(iphone11)}")
iphone11_min, iphone11_max, min_post_iphone11 = check_price_range(iphone11, 200, 2000)
print(f"iPhone 11 Price Range: {iphone11_min}$ - {iphone11_max}$\n")
print("Min post link:")
print(min_post_iphone11["Post Link"])
print()

print(f"iPhone 10: {len(iphone10)}")
iphone10_min, iphone10_max, min_post_iphone10 = check_price_range(iphone10, 100, 2000)
print(f"iPhone 10 Price Range: {iphone10_min}$ - {iphone10_max}$\n")
print("Min post link:")
print(min_post_iphone10["Post Link"])
print()

print(f"iPad: {len(ipad)}")
ipad_min, ipad_max, min_post_ipad = check_price_range(ipad, 200, 2000)
print(f"iPad Price Range: {ipad_min}$ - {ipad_max}$\n")
print("Min post link:")
print(min_post_ipad["Post Link"])
print()

print(f"Below 10 series: {len(below_ten)}")
below_ten_min, below_ten_max, min_post_below_ten = check_price_range(below_ten, 100, 2000)
print(f"Below 10 series Price Range: {below_ten_min}$ - {below_ten_max}$\n")
print("Min post link:")
print(min_post_below_ten["Post Link"])
print()
```

```python
samsung_data_file = get_today_file_path(constants.SMARTPHONES[1]["file"])
samsung_data = read_from_text_to_dicts_list(samsung_data_file)


def check_samsung_series():

    s22 = []
    s22_ultra = []
    s22_plus = []

    s21 = []
    s21_ultra = []
    s21_plus = []

    s20 = []
    s20_ultra = []
    s20_plus = []

    fold = []

    for i in samsung_data:

        if (
            "oct" in i["Post Date"].lower()
            or "ago" in i["Post Date"].lower()
            or "yesterday" in i["Post Date"].lower()
        ):  # If post is within October or recently

            if (
                "s22" in i["Name"].lower()
                or "s22" in i["Description"].lower()
                or "s 22" in i["Description"].lower()
                or "s 22" in i["Name"].lower()
            ):
                if "ultra" in i["Name"].lower() or "ultra" in i["Description"].lower():
                    s22_ultra.append(i)
                elif (
                    "plus" in i["Name"].lower()
                    or "plus" in i["Description"].lower()
                    or "+" in i["Name"].lower()
                    or "+" in i["Description"].lower()
                ):
                    s22_plus.append(i)
                else:
                    s22.append(i)

            elif (
                "s21" in i["Name"].lower()
                or "s21" in i["Description"].lower()
                or "s 21" in i["Description"].lower()
                or "s 21" in i["Name"].lower()
            ):
                if "ultra" in i["Name"].lower() or "ultra" in i["Description"].lower():
                    s21_ultra.append(i)
                elif (
                    "plus" in i["Name"].lower()
                    or "plus" in i["Description"].lower()
                    or "+" in i["Name"].lower()
                    or "+" in i["Description"].lower()
                ):
                    s21_plus.append(i)
                else:
                    s21.append(i)

            elif (
                "s20" in i["Name"].lower()
                or "s20" in i["Description"].lower()
                or "s 20" in i["Description"].lower()
                or "s 20" in i["Name"].lower()
            ):
                if "ultra" in i["Name"].lower() or "ultra" in i["Description"].lower():
                    s20_ultra.append(i)
                elif (
                    "plus" in i["Name"].lower()
                    or "plus" in i["Description"].lower()
                    or "+" in i["Name"].lower()
                    or "+" in i["Description"].lower()
                ):
                    s20_plus.append(i)
                else:
                    s20.append(i)
            elif(
                'fold' in i['Name'].lower()
                or 'fold' in i['Description'].lower()
            ):
                fold.append(i)


    print(f"S22: {len(s22)}")
    s22_min, s22_max, min_s22_post = check_price_range(s22, 200, 2500)
    print(f"S22 Price Range: {s22_min} - {s22_max}$")
    print(f"S22 Min price post: {min_s22_post['Post Link']}\n")

    print(f"S22 Plus: {len(s22_plus)}")
    s22_plus_min, s22_plus_max, min_s22_plus_post = check_price_range(
        s22_plus, 200, 2500
    )
    print(f"S22 Price Range: {s22_plus_min} - {s22_plus_max}$")
    print(f"S22 Min price post: {min_s22_plus_post['Post Link']}\n")

    print(f"S22 Ultra: {len(s22_ultra)}")
    s22_ultra_min, s22_ultra_max, min_s22_ultra_post = check_price_range(
        s22_ultra, 400, 2500
    )
    print(f"S22 Price Range: {s22_ultra_min} - {s22_ultra_max}$")
    print(f"S22 Min price post: {min_s22_ultra_post['Post Link']}\n")


    samsung = [
        {
            "s22": [s22, s22_ultra, s22_plus],
            "s21": [s21, s21_ultra, s21_plus],
            "s20": [s20, s20_ultra, s20_plus],
            "fold": fold
        }
    ]

    return samsung


check_samsung_series()
```
